```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.analyticsvidhya.com/blog/2023/01/top-100-data-science-interview-questions/"
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    html_content = response.content
else:
    print(f"Failed to fetch the page: {response.status_code}")
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
questions_answers = []

# Find all question elements
for question in soup.find_all('h4'):
    question_text = question.get_text(strip=True)
    answer = question.find_next_sibling('p')
    if answer:
        answer_text = answer.get_text(strip=True)
        questions_answers.append({"question": question_text, "answer": answer_text})

# Print extracted data
for qa in questions_answers:
    print(qa)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import gradio as gr

# Load the SentenceTransformer model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Dataset of questions and answers
qa_data =  [
    {"question": "Which is faster, Python list or Numpy arrays, and why?",
     "answer": "NumPy arrays are faster than Python lists for numerical operations because they use optimized C libraries."},
    {"question": "What is the difference between a Python list and a tuple?",
     "answer": "Lists are mutable (can be changed) while tuples are immutable (cannot be changed)."},
    {"question": "What is the use of the pass keyword in Python?",
     "answer": "The 'pass' keyword is a placeholder for code that hasn't been written yet."},
    {"question": "Q4. What is the difference between split and join?",
     "answer": "A Split and join are both functions of python strings, but they are completely different when it comes to functioning."},
    {"question": "Q5. Explain the logical operations in python.",
     "answer": "A. In Python, the logical operations and, or, and not can be used to perform boolean operations on truth values (True and False)."},
    {"question": "Q6. Explain the top 5 functions used for python strings.",
     "answer": "A. Here are the top 5 Python string functions."},
    {"question": "Q7. What is the use of the pass keyword in python?",
     "answer": "A. The pass keyword is a null statement that does nothing. It is often used as a placeholder where a statement is required syntactically, but no action needs to be taken."},
    {"question": "Q8. What is the use of the continue keyword in python?",
     "answer": "A. continue is used in a loop to skip over the current iteration and move on to the next one."},
    {"question": "Q9. What are immutable and mutable data types?",
     "answer": "A. In Python, immutable objects cannot be modified after creation (e.g., numbers, strings, tuples), while mutable objects can (e.g., lists, dictionaries)."},
    {"question": "Q10. What is the use of try and except block in python?",
     "answer": "A. The try and except block in Python are used to handle exceptions that occur during program execution."},
    {"question": "Q21. How to perform univariate analysis for numerical and categorical variables?",
     "answer": "A. Univariate analysis is used to analyze and describe the characteristics of a single variable. It involves understanding distribution, central tendency, and dispersion for numerical data, and frequencies for categorical data."},
    {"question": "Q22. What are the different ways in which we can find outliers in the data?",
     "answer": "A. Methods to find outliers include using box plots, Z-scores, and the interquartile range (IQR)."},
    {"question": "Q23. What are the different ways by which you can impute the missing values in the dataset?",
     "answer": "A. Missing values can be imputed using techniques like mean, median, or mode imputation, or by using predictive modeling."},
    {"question": "Q24. What is skewness in statistics and its types?",
     "answer": "A. Skewness measures the asymmetry of a distribution. Positive skewness has a longer tail on the right, while negative skewness has a longer tail on the left."},
    {"question": "Q25. What are the measures of central tendency?",
     "answer": "A. Measures of central tendency include mean, median, and mode, which describe the center of a dataset."},
    {"question": "Q26. Can you explain the difference between descriptive and inferential statistics?",
     "answer": "A. Descriptive statistics summarize a dataset using measures like mean and standard deviation, while inferential statistics make predictions or inferences about a population based on sample data."},
    {"question": "Q27. What are the key elements of an EDA report?",
     "answer": "A. Key elements include univariate analysis, bivariate analysis, missing data analysis, and visualizations to understand data patterns."},
    {"question": "Q28. What is the central limit theorem?",
     "answer": "A. The central limit theorem states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population distribution."},
    {"question": "Q29. Mention the two kinds of target variables for predictive modeling.",
     "answer": "A. The two kinds are categorical (classification) and numerical (regression)."},
    {"question": "Q30. When will the mean, median, and mode be the same for a dataset?",
     "answer": "A. The mean, median, and mode will be the same if the dataset follows a perfectly normal distribution."},
    {"question": "Q31. What is the difference between Variance and Bias in Statistics?",
     "answer": "A. Variance measures the model's sensitivity to small changes in data, while bias measures error due to overly simplistic assumptions in the model."},
    {"question": "Q32. What is the difference between Type I and Type II errors?",
     "answer": "A. Type I error occurs when a true null hypothesis is rejected, while Type II error occurs when a false null hypothesis is not rejected."},
    {"question": "Q33. What is the Confidence Interval in statistics?",
     "answer": "A. A confidence interval is a range of values within which a population parameter is expected to lie, with a certain level of confidence (e.g., 95%)."},
    {"question": "Q34. Can you explain the concept of correlation and covariance?",
     "answer": "A. Correlation measures the strength and direction of the relationship between two variables, while covariance indicates how two variables change together."},
    {"question": "Q35. Why is hypothesis testing useful for a data scientist?",
     "answer": "A. Hypothesis testing helps validate claims or assumptions about data and supports decision-making based on statistical evidence."},
    {"question": "Q36. What is a chi-square test of independence used for in statistics?",
     "answer": "A. A chi-square test determines if there is a significant association between two categorical variables."},
    {"question": "Q37. What is the significance of the p-value?",
     "answer": "A. The p-value indicates the probability of observing the data, or something more extreme, under the null hypothesis. A low p-value suggests strong evidence against the null hypothesis."},
    {"question": "Q38. What are the different types of sampling techniques?",
     "answer": "A. Sampling techniques include random sampling, stratified sampling, cluster sampling, and systematic sampling."},
    {"question": "Q39. What is Bayes' theorem and how is it used in data science?",
     "answer": "A. Bayes' theorem calculates conditional probabilities. It is widely used in machine learning for classification tasks."},
    {"question": "Q40. What is the difference between parametric and non-parametric tests?",
     "answer": "A. Parametric tests assume underlying data distribution (e.g., t-tests), while non-parametric tests do not (e.g., Mann-Whitney U test)."}
]

qa_data.append({
    "question": "What is Python?",
    "answer": "Python is the fastest-growing programming language. It is a set of instructions given in the form of a program to a computer to perform specific tasks. Python is an interpreted, object-oriented, and high-level programming language. Due to its beginner-friendly syntax, it has become a clear choice for beginners starting their programming journey."
})


# Precompute embeddings for all questions
question_embeddings = [model.encode(q['question']) for q in qa_data]

# Function to calculate cosine similarity
def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

# Chatbot function
def chatbot(user_input):
    user_embedding = model.encode(user_input)  # Encode the user's input
    similarities = [cosine_similarity(user_embedding, q_emb) for q_emb in question_embeddings]

    # Find the best match
    best_match_idx = np.argmax(similarities)
    best_similarity = similarities[best_match_idx]

    # Define a threshold for similarity
    if best_similarity > 0.7:  # You can adjust this threshold
        return qa_data[best_match_idx]['answer']
    else:
        return "I'm sorry, I couldn't find an answer to your question. Please try rephrasing it."

# Gradio interface
interface = gr.Interface(
    fn=chatbot,
    inputs="text",
    outputs="text",
    title="Data Science Chatbot",
    description="Ask me any data science-related question! The chatbot will try to provide the best answer."
)

# Launch the chatbot
interface.launch()
```
